Lo que vamos a hacer es calcular distancias entre puntos de aforament por las calles de Barcelona.

Para calcular los kilómetros recorridos de forma realista no se ha utilizado distancia euclidiana, sino distancia sobre red viaria. Para ello se ha descargado la red de conducción de Barcelona desde OpenStreetMap mediante OSMnx y se ha aplicado el algoritmo de Dijkstra para obtener la distancia mínima entre nodos asociada a cada equipo de aforo.

En el siguiente paso lo que hacemos es descargar:

* Todas las calles transitables en coche

* En formato grafo (nodos e intersecciones)


Un grafo tiene:

* Nodos (intersecciones)

* Aristas (tramos de calle)

* Cada arista tiene una longitud real en metros


```python
import osmnx as ox

# Descargar red de conducción (drive). Lo que hacemos es descargar las red viaria de las calles de Barcelona en modo grafo asignado una nodo y aristas a cada calle
red_conduccion = ox.graph_from_place("Barcelona, Spain", network_type="drive")
```

```python
red_conduccion
```

Convertir equipos en nodos de red:

Cada equipo debe asociarse al nodo más cercano de la red viaria que nos hemos descargado con osmnx. Porque el cálculo de ruta se hace entre nodos del grafo.

* Mira la coordenada de tu equipo de aforo.

* Busca cuál es la intersección de calle más cercana.

* Devuelve el ID de esa intersección.


Es decir el equipo no está exactamente en un nodo del grafo, entonces se "engancha" al nodo más cercano para poder calcular las rutas. y el nearest_node es el id que se le asigna al equipo para después calcular las distancias


```python
# Obtener nodos más cercanos
geo_equips_aforats_sin_bici_bcn["nearest_node"] = geo_equips_aforats_sin_bici_bcn.apply(
    lambda row: ox.distance.nearest_nodes(red_conduccion, row["Longitud"], row["Latitud"]),
    axis=1
)
```

```python
geo_equips_aforats_sin_bici_bcn
```

Calcular distancia al equipo más cercano



```python
import networkx as nx

distancias_minimas = []

nodos = geo_equips_aforats_sin_bici_bcn["nearest_node"].tolist()

for origen in nodos:
    
    # Calcula TODAS las distancias desde este origen
    lengths = nx.single_source_dijkstra_path_length(
        red_conduccion,
        origen,
        weight="length"
    )
    
    # Filtramos solo nodos que sean equipos de aforo
    distancias_a_equipos = [
        lengths[nodo] 
        for nodo in nodos 
        if nodo in lengths and nodo != origen
    ]
    
    if distancias_a_equipos:
        distancias_minimas.append(min(distancias_a_equipos))
    else:
        distancias_minimas.append(None)

geo_equips_aforats_sin_bici_bcn["distancia_equipo_mas_cercano_m"] = distancias_minimas
```

```python
geo_equips_aforats_sin_bici_bcn
```

Las distancias se han calculado utilizando la red viaria descargada mediante OSMnx, donde cada arista incorpora su longitud real en metros. El algoritmo de Dijkstra ponderado por el atributo "length" permite obtener la distancia real recorrida por carretera entre puntos.

```python
geo_equips_aforats_sin_bici_bcn["distancia_km"] = (
    geo_equips_aforats_sin_bici_bcn["distancia_equipo_mas_cercano_m"] / 1000
)
```
